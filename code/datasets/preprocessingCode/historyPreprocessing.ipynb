```python
import pandas as pd
import numpy as np

with open("../unprocessed/history.txt", errors="replace") as f:
    lines = np.asarray(f.readlines())

print(len([s for s in lines if s[0] == "#"]))

for line in lines:
    print(line.rstrip())
```

```python
history = pd.DataFrame({
    "question": np.asarray([], dtype=object),
    "answer": np.asarray([], dtype=object),
    "qA": np.asarray([], dtype=object)
})

i = 1
while i < len(lines):
    if lines[i][0] == "#":
        row = {
            "question": lines[i], 
            "answer": lines[i+1].strip("^ \n"),
            "qA": (lines[i:i+6])
        }
        history.loc[len(history)] = row
    i += 1

history
```

```python
history[:500].to_csv("../processed/score/history.csv", index = False)
history[500:].to_csv("../processed/evaluation/history.csv", index = False)
```
